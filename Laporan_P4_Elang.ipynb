<p align="justify"> <br>


*   Positional Prameters <br>

```
void cetakInfo(String nama, int usia) {
  print('Nama: $nama, Usia: $usia');
}
```
*   Named Parameters  <br>

```
void cetakInfo({String nama, int usia}) {
  print('Nama: $nama, Usia: $usia');
}
```
*   Default Parameters <br>

```
void cetakInfo({String nama = 'Tidak Ada Nama', int usia = 0}) {
  print('Nama: $nama, Usia: $usia');
}
```
*   Required Parameters  <br>

```
void cetakInfo({required String nama, required int usia}) {
  print('Nama: $nama, Usia: $usia');
}
```
*   Rest Parameters  <br>

```
void tampilkanHobi(String nama, String pekerjaan, {required String... hobi}) {
  print('Nama: $nama, Pekerjaan: $pekerjaan');
  print('Hobi: $hobi');
}
```

4. Jelaskan maksud Functions sebagai first-class objects beserta contoh sintaknya!

Jawab:

<p align="justify"> Function di dalam Dart dapat di passing ke parameter dari function lainnya.

```
void main(List<String> args) {
  //Function yang dipassing ke parameter function lainnya.
  void printElement(int element) => print(element);
  //Function yang di assign ke variabel
  final filter = (int element) => element.isOdd;
  List.generate(100, (index) => index + 1).where(filter).forEach(printElement);
}
```

5. Apa itu Anonymous Functions? Jelaskan dan berikan contohnya!

Jawab:

<p align="justify"> Sebuah function yang tidak memiliki nama, biasanya function mempunyai nama, contoh main() dimana main merupakan namanya. Anonymous function biasa dikenal dengan lamba atau closure di dalam bahasa pemrograman lain.

```
void main(List<String> args) {
  List.generate(100, (index) => index + 1)
      .where((int item) => item.isEven)
      .forEach((int element) => print('Value: $element'));
}
```

6. Jelaskan perbedaan Lexical scope dan Lexical closures! Berikan contohnya!

Jawab:

<p align="justify"> Dart adalah bahasa yang dibatasi leksikal, yang berarti bahwa ruang lingkup variabel ditentukan secara statis, hanya dengan letak kode (lexical scope). Sedangkan Closure adalah sebuah function yang memiliki akses ke variabel dalam lingkup leksikalnya, bahkan fungsi tersebut digunakan di luar lingkup aslinya.


*   Lexical Scope  <br>

```
bool topLevel = true;

void main(List<String> args) {
  bool insideMain = topLevel;

  void function() {
    bool insideFunction = insideMain;
  }
}
```
*   Lexical Closures  <br>

```
Function addWithTen(int number) => (int amount) => number + amount + 10;

void main(List<String> args) {
  final count = addWithTen(1);
  final digit = addWithTen(2);

  print([count(1), digit(2)]);
  //Output: [12, 14]
}
```

7. Jelaskan dengan contoh cara membuat return multiple value di Functions!

Jawab:

<p align="justify"> Dalam bahasa pemrograman Dart, Anda dapat mengembalikan multiple value dari sebuah fungsi dengan menggabungkannya dalam sebuah objek atau tuple. Cara yang paling umum adalah menggunakan objek Map atau objek yang ditentukan sendiri.

```
void main() {
  Tuple3<String, int, bool> getPersonInfo() {
    String name = "John";
    int age = 30;
    bool isStudent = false;

    return Tuple3(name, age, isStudent);
  }

  // Memanggil fungsi getPersonInfo()
  Tuple3<String, int, bool> personInfo = getPersonInfo();

  // Mengakses multiple value yang dikembalikan
  String name = personInfo.item1;
  int age = personInfo.item2;
  bool isStudent = personInfo.item3;

  print("Name: $name");
  print("Age: $age");
  print("Is Student: $isStudent");
}
```

<p align="justify"> Pada contoh diatas, kita menggunakan pustaka eksternal seperti tuple untuk mendefinisikan tipe data tuple dan mengembalikannya dari fungsi. Kemudian, kita dapat mengakses nilai-nilai tuple menggunakan item1, item2, dan item3.
